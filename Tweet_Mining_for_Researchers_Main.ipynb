```python
import re
import html

tweet_shortcuts = {
    "u": "you",
    "r": "are",
    "ur": "your",
    "4": "for",
    "2": "to",
    "b4": "before",
    "gr8": "great",
    "thru": "through",
    "bcoz": "because",
    "l8r": "later",
    "pls": "please",
    "ppl": "people",
    "y": "why",
    "cuz": "because",
    "idk": "i don't know",
    "im": "i am",
    "omg": "oh my god",
    "lol": "laughing out loud",
    "brb": "be right back",
    "btw": "by the way",
    "gtg": "got to go",
    "g2g": "got to go",
    "ttyl": "talk to you later",
    "np": "no problem",
    "ty": "thank you",
    "thx": "thanks",
    "w/": "with",
    "w/o": "without"
}

def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                               u"\U0001F700-\U0001F77F"  # Alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002500-\U00002BEF"  # Chinese characters and misc symbols
                               u"\U0001F004-\U0001F0CF"  # Mahjong and Playing cards
                               u"\U00002702-\U000027B0"  # Dingbats
                               "]+", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)

def replace_shortcuts(text):
    """Replaces common tweet shortcuts with their full forms."""
    words = text.split()
    for i, word in enumerate(words):
        if word in tweet_shortcuts:
            words[i] = tweet_shortcuts[word]
    return " ".join(words)

def extract_mentions(text):
    """Extract mentions from the tweet text."""
    return re.findall(r"@\w+", text)

def extract_hashtags(text):
    """Extract hashtags from the tweet text."""
    return re.findall(r"#\w+", text)

def extract_urls(text):
    """Extract URLs from the tweet text."""
    return re.findall(r"https?://\S+", text)

def preprocess_tweet(tweet_text):
    # Remove emojis
    tweet_text = remove_emojis(tweet_text)

    # Convert to lowercase
    tweet_text = tweet_text.lower()

    # Remove newline characters
    tweet_text = tweet_text.replace("\\n", " ").replace("\n", " ")

    # Replace common tweet shortcuts
    tweet_text = replace_shortcuts(tweet_text)

    # Remove HTML escape sequences (if any remain)
    tweet_text = re.sub(r"&\w+;", "", tweet_text)

    # Remove unwanted punctuations but keep URLs & hashtags
    tweet_text = re.sub(r"[^\w\s#@:/.]", "", tweet_text)

    return tweet_text

def process_tweets_file(file_path):
    """Reads tweets from a text file and processes them."""
    tweet_data = []

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

            # Normalize formatting issues
            content = content.replace("'>,", "'>\n").replace("<Tweet", "\n<Tweet")

            for line in content.split('\n'):
                line = line.strip()
                if not line.startswith('<Tweet'):
                    continue

                # Extract tweet ID and text
                match = re.search(r"<Tweet id=(\d+) text=['\"](.*?)['\"]>", line)
                if not match:
                    continue

                tweet_id, tweet_text = match.groups()
                tweet_text = preprocess_tweet(tweet_text)

                mentions = extract_mentions(tweet_text)
                hashtags = extract_hashtags(tweet_text)
                urls = extract_urls(tweet_text)

                tweet_entry = {
                    "tweet_id": tweet_id,
                    "text": tweet_text,
                    "mentions": mentions,
                    "hashtags": hashtags,
                    "urls": urls
                }
                tweet_data.append(tweet_entry)

        if tweet_data:
            return tweet_data
        else:
            print("No tweets found in the file.")

    except Exception as e:
        print(f"Error processing file: {e}")
        import traceback
        traceback.print_exc()

file_path = "/content/drive/MyDrive/SEM 8/Capstone/tweets.txt"
# file_path='tweets.txt'
process_tweets_file(file_path)
```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
# Define the prompt template for the LLM
prompt = ChatPromptTemplate.from_template(
    template="""You are an assistant for question-answering tasks.
    Use the following documents to answer the question.
    If you don't know the answer, just say that you don't know.
    Use three sentences maximum and keep the answer concise:
    Question: {question}
    Documents: {context}
    Answer:
    """
)
```

#**RAG Chain**

```python
rag_chain = prompt | llm | StrOutputParser()
```

```python
class RAGApplication:
    def __init__(self, retriever, rag_chain):
        self.retriever = retriever
        self.rag_chain = rag_chain
    def run(self, question):
        # Retrieve relevant documents
        documents = self.retriever.invoke(question)
        # Extract content from retrieved documents
        doc_texts = "\\n".join([doc.page_content for doc in documents])
        # Get the answer from the language model
        answer = self.rag_chain.invoke({"question": question, "documents": doc_texts})
        return answer
```

```python
from langchain.schema.runnable import RunnablePassthrough
ragchain = (
    {'context':retriever,"question":RunnablePassthrough()}
    | prompt
)
ragchain.invoke("Are there any tweets related to having breastcancer?")
```

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain.chains import create_history_aware_retriever
from langchain.chains.combine_documents import create_stuff_documents_chain

contextualize_q_system_prompt = """
Given a chat history and the latest user question
which might reference context in the chat history,
formulate a standalone question which can be understood
without the chat history. Do NOT answer the question,
just reformulate it if needed and otherwise return it as is.
"""

contextualize_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", contextualize_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)

contextualize_chain = contextualize_q_prompt | llm | StrOutputParser()
print(contextualize_chain.invoke({"input": "Give more explanation on it ", "chat_history": chat_history}))
```

```python
from langchain.chains import create_retrieval_chain,history_aware_retriever
from langchain.chains.combine_documents import create_stuff_documents_chain

history_aware_retriever = create_history_aware_retriever(
    llm, retriever, contextualize_q_prompt
)

qa_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant. Use the following context to answer the user's question."),
    ("system", "Context: {context}"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

question_answer_chain = create_stuff_documents_chain(llm, qa_prompt)
rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)

```

```python
import sqlite3
from datetime import datetime

DB_NAME = "tweets.db"

def get_db_connection():
  conn = sqlite3.connect(DB_NAME)
  conn.row_factory = sqlite3.Row
  return conn

def create_application_logs():
  conn = get_db_connection()
  conn.execute(''' CREATE TABLE IF NOT EXISTS application_logs
                        (ID INTEGER PRIMARY KEY AUTOINCREMENT,
                        session_id TEXT,
                        user_query TEXT,
                        llama_response TEXT,
                        model TEXT,
                        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP)''')
  conn.close()

def insert_application_logs(session_id, user_query, llama_response, model):
    conn = get_db_connection()
    conn.execute('INSERT INTO application_logs (session_id, user_query, llama_response, model) VALUES (?, ?, ?, ?)',
                 (session_id, user_query, llama_response, model))
    conn.commit()
    conn.close()

def get_chat_history(session_id):
    conn = get_db_connection()
    cursor = conn.cursor()
    cursor.execute('SELECT user_query, llama_response FROM application_logs WHERE session_id = ? ORDER BY created_at', (session_id,))
    messages = []
    for row in cursor.fetchall():
        messages.extend([
            {"role": "human", "content": row['user_query']},
            {"role": "ai", "content": row['llama_response']}
        ])
    conn.close()
    return messages

def drop_application_logs():
    conn = get_db_connection()
    conn.execute('DROP TABLE IF EXISTS application_logs')
    conn.commit()
    conn.close()
    print("Application logs table dropped successfully!")

# Call the function to drop the table
drop_application_logs()
create_application_logs()

import uuid
session_id = str(uuid.uuid4())
chat_history = get_chat_history(session_id)
print(chat_history)
question1 = "Are there any tweets related to having breastcancer?"
response1 = rag_chain.invoke({"input": question1, "chat_history": chat_history})['answer']
insert_application_logs(session_id, question1, response1, "llama3")
print(f"Human: {question1}")
print(f"AI: {response1}\n")
```
