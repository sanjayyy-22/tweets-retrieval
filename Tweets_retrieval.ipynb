```python
from pymongo import MongoClient
from pymongo.server_api import ServerApi
import re

def process_tweets_file(file_path):
    """Reads tweets from a text file and stores them in MongoDB with original text."""
    tweet_data = []

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

            # Convert the string representation of the list to a single string
            # This makes it easier to process with regex
            content = content.replace("'>,", "'>\n").replace("<Tweet", "\n<Tweet")

            for line in content.split('\n'):
                line = line.strip()
                if not line or not line.startswith('<Tweet'):
                    continue

                # Extract ID from the beginning
                id_match = re.search(r'<Tweet id=(\d+)', line)
                if not id_match:
                    continue
                tweet_id = id_match.group(1)

                # Extract text content between first occurrence of 'text=' and the final '>'
                text_start = line.find('text=') + 5  # +5 to get past 'text='
                if text_start < 5:  # 'text=' not found
                    continue

                # Determine which quote character is used (single or double)
                quote_char = line[text_start]
                if quote_char not in ["'", '"']:
                    continue

                # Find the matching closing quote by working backwards from the end
                text_end = line.rfind(quote_char + '>')
                if text_end < 0:
                    text_end = line.rfind(quote_char)

                if text_end < text_start:
                    continue

                # Extract the original tweet text without any cleaning or processing
                tweet_text = line[text_start+1:text_end]

                # Create tweet entry with original text
                tweet_entry = {
                    "tweet_id": tweet_id,
                    "text": tweet_text
                }
                tweet_data.append(tweet_entry)

        # Insert into MongoDB if we have tweets
        if tweet_data:
            #collection.insert_many(tweet_data)
            #print(f"{len(tweet_data)} tweets inserted successfully!")
            return tweet_data
        else:
            print("No tweets found in the file.")

    except Exception as e:
        print(f"Error processing file: {e}")
        import traceback
        traceback.print_exc()

# Process tweets from the file
#file_path = "/content/drive/MyDrive/SEM 8/Alternative Data/tweets.txt"
file_path='tweets.txt'
process_tweets_file(file_path)
```
